```
#讀取opendata資料 純到Petdata
Petdata<-jsonlite::fromJSON("http://data.taipei/opendata/datalist/apiAccess?scope=resourceAquire&rid=f4a75ba9-7721-4363-884d-c3820b0b917c")
str(Petdata)
```

```
#擷取results內容
head(Petdata$result$results)
```
